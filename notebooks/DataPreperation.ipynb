let's merge credits and keywords to the movies dataset

```python
movies_df = movies_df[movies_df["id"].apply(lambda x: "-" not in str(x))]
movies_df["id"] = movies_df["id"].astype(int)
movies_df = pd.merge(movies_df, credits_df, on="id", how="left")
movies_df = pd.merge(movies_df, keywords_df, on="id", how="left")
```

```python
movies_df.head()
```

```python
movies_df.isnull().sum()
```

Let's deal with the missing values

```python
movies_df.columns
```

```python
columns_to_drop = ["belongs_to_collection", "homepage", "status", "video"]
movies_df = movies_df.drop(columns_to_drop, axis=1)
```

```python
text_columns = ["overview", "tagline", "original_language", "poster_path"]
movies_df[text_columns] = movies_df[text_columns].fillna("DNE")
```

```python
numbers_columns = ["runtime"]
movies_df[numbers_columns] = movies_df[numbers_columns].fillna(0)

movies_df.dropna(inplace=True)
movies_df.isnull().sum()
```

```python
from datetime import datetime

movies_df["year"] = pd.to_datetime(movies_df["release_date"], errors="coerce").dt.year
movies_df["year"] = movies_df["year"].astype(int)

movies_df["release_date"] = pd.to_datetime(
    movies_df["release_date"], format="%Y-%m-%d", errors="coerce"
)

movies_df["popularity"] = movies_df["popularity"].astype("float64")

ratings_df["timestamp"] = ratings_df["timestamp"].apply(
    lambda x: datetime.fromtimestamp(x)
)
```

```python
import ast
from Sameer.DataPreperation import (
    extend_list_from_column,
    get_director,
    get_list,
    clean_text_data,
    create_bag_of_words,
)

spoken_languages_list = []
cast_list = []
crew_list = []
company_list = []
country_list = []
original_language_list = []
```

```python
for lang in movies_df["original_language"]:
    original_language_list.extend(lang.split(", "))

extend_list_from_column(movies_df, "spoken_languages", spoken_languages_list)
extend_list_from_column(movies_df, "cast", cast_list, key="character")
extend_list_from_column(movies_df, "crew", crew_list)
extend_list_from_column(movies_df, "production_companies", company_list)
extend_list_from_column(movies_df, "production_countries", country_list)
```

```python
for col in ["crew", "cast", "genres", "keywords", "production_companies"]:
    movies_df[col] = movies_df[col].fillna("[]").apply(ast.literal_eval)

movies_df["director"] = movies_df["crew"].apply(get_director)

for col in ["cast", "genres", "keywords", "production_companies"]:
    movies_df[col] = movies_df[col].apply(get_list)

features = ["adult", "cast", "keywords", "director", "genres", "production_companies"]
for feature in features:
    movies_df[feature] = movies_df[feature].apply(clean_text_data)

movies_df["bag_of_words"] = movies_df.apply(create_bag_of_words, axis=1)
```
